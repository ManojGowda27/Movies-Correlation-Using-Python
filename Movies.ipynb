```python
# Reading the Data
df = pd.read_csv("/content/movies.csv")
```

```python
# Displaying the Data

df.head()
```

```python
# Finding the missing Data

for col in df.columns:
 pct_missing = np.mean(df[col].isnull())
 print('{} - {}%'.format(col, pct_missing))
```

```python
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Removing Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

**Assumption**

Budget and Company has High Correlation with Gross Earnings

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget')
plt.show()
```

```python
df.head()
```

```python
# Regression Plot
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Displaying Correlation using different correlation methods

df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
correlation_matrrix = df.corr(method='pearson')
sns.heatmap(correlation_matrrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
  if(df_numerized[col_name].dtype == 'object'):
    df_numerized[col_name] = df_numerized[col_name].astype('category')
    df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df
```

```python
correlation_matrrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrrix, annot=True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Unstacking
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

**Conclusion**

Votes and Budget have the highest correlation to gross earnings as compared to Budget and Company.
